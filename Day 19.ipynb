```python
import numpy as np
from numpy.linalg import inv

rotations = [
    np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]]),
    np.array([[0, 0, 1], [0, 1, 0], [-1, 0, 0]]),
    np.array([[-1, 0, 0], [0, 1, 0], [0, 0, -1]]),
    np.array([[0, 0, -1], [0, 1, 0], [1, 0, 0]]),
    np.array([[0, -1, 0], [1, 0, 0], [0, 0, 1]]),
    np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]),
    np.array([[0, 1, 0], [1, 0, 0], [0, 0, -1]]),
    np.array([[0, 0, -1], [1, 0, 0], [0, -1, 0]]),
    np.array([[0, 1, 0], [-1, 0, 0], [0, 0, 1]]),
    np.array([[0, 0, 1], [-1, 0, 0], [0, -1, 0]]),
    np.array([[0, -1, 0], [-1, 0, 0], [0, 0, -1]]),
    np.array([[0, 0, -1], [-1, 0, 0], [0, 1, 0]]),
    np.array([[1, 0, 0], [0, 0, -1], [0, 1, 0]]),
    np.array([[0, 1, 0], [0, 0, -1], [-1, 0, 0]]),
    np.array([[-1, 0, 0], [0, 0, -1], [0, -1, 0]]),
    np.array([[0, -1, 0], [0, 0, -1], [1, 0, 0]]),
    np.array([[1, 0, 0], [0, -1, 0], [0, 0, -1]]),
    np.array([[0, 0, -1], [0, -1, 0], [-1, 0, 0]]),
    np.array([[-1, 0, 0], [0, -1, 0], [0, 0, 1]]),
    np.array([[0, 0, 1], [0, -1, 0], [1, 0, 0]]),
    np.array([[1, 0, 0], [0, 0, 1], [0, -1, 0]]),
    np.array([[0, -1, 0], [0, 0, 1], [-1, 0, 0]]),
    np.array([[-1, 0, 0], [0, 0, 1], [0, 1, 0]]),
    np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]),
    ]

inv_rotations = [inv(x) for x in rotations]

def rotate(matrix):
    '''Returns all 24 rotations of a set of coordinates.'''
    return [np.transpose(np.matmul(x, np.transpose(matrix))) for x in rotations]
```

```python
def is_pair(a1, a2):
    '''Check whether two scanners have at least 12 beacons in common. If they do, they are a pair.
    It also returns the respective beacon pair indexes (scanner1, scanner2).
    It also returns the relative position shift of scanner 2 from scanner 1.
    This already assumes the orientation of the two scanners are the same.'''
    deltas = []
    deltas_dic = defaultdict(list)
    
    for i in range(len(a1)): # For each beacon in scanner 1,
        for j in range(len(a2)): # find the distance to each beacon in scanner 2.
            deltas_dic[(i,j)] = list(a1[i]-a2[j]) # Add the distance of each beacon pair to a dictionary.
            deltas.append(list(a1[i]-a2[j])) # Add the distance to a list.
    
    
    count = defaultdict(int)
    for dist in deltas:
        count[str(dist)] += 1 # Summing up the counts of each distance. Need to do str(dist) for hashable.
    
    
    if count[max(count, key=count.get)] >= 12: # If more than 12 distances are the same, then they are a pair, and can be used to map out the area.
        return True, [k for k,v in deltas_dic.items() if str(v)==max(count, key=count.get)], max(count, key=count.get)  # Add the common (beacon1, beacon2) pair indexes.
    else:
        return False
```

```python
from math import ceil
import ast
def get_pairs(dictionary):
    '''Get all the scanner pairs by fixing a scanner, and going through each other scanner and rotating
    them to find whether they are a pair. Also returns the rotation number and relative position
    of scanner2 from scanner1 for pairs = (scanner1, scanner2).
    
    Returns (pairs, rotation number for pairs, and relative positions)
    
    Function can also provide the common beacons, but is not necessary.'''
    
    pairs = []
    relative_positions = []
    #common_beacons = defaultdict(list)
    rotation_num = []
    
    n_scanners = len(dictionary)
    
    for scanner in range(ceil(n_scanners)):
        for possible_pair in range(n_scanners):
            if scanner != possible_pair: # Make sure it's not the same scanner
                rotations = rotate(dictionary[possible_pair])
                for rotation_index in range(len(rotations)): # Finding the correct orientation of the possible pair.
                    if (is_pair(dictionary[scanner], rotations[rotation_index])) and ((possible_pair, scanner) not in pairs): # Add pairs, and also not double count.
                        pairs.append((scanner, possible_pair))
                        rotation_num.append(rotation_index)
                        relative_positions.append(is_pair(dictionary[scanner], rotations[rotation_index])[2])
                        #common_beacons[(scanner, possible_pair)] = is_pair(dictionary[scanner], rotations[rotation_index])[1]
    return pairs, rotation_num, [np.array(ast.literal_eval(x)) for x in relative_positions]#,  common_beacons
```

```python
def orient(dictionary):
    '''The transformation required for every scanner to be the same as scanner 0.
    Returns the initial dictionary with all coordinates in the correct orientation.'''
    orient_transform = defaultdict(int)
    not_oriented = list(range(1,len(dictionary)))
    
    pair_info = get_pairs(dictionary)
    
    
    # Initial conditions (Scanner pairs with 0)
    orient_transform[0] = rotations[0]
    for i in range(len(get_pairs(dictionary)[0])):
        if pair_info[0][i][0] == 0:
            not_oriented.remove(pair_info[0][i][1])
            orient_transform[pair_info[0][i][1]] = rotations[pair_info[1][i]]
    
    
    while len(not_oriented) != 0:
         for scanner in not_oriented:
            for i in range(len(pair_info[0])):
                if (scanner == pair_info[0][i][1]) and (pair_info[0][i][0] in orient_transform):
                    orient_transform[scanner] = np.matmul(orient_transform[pair_info[0][i][0]], rotations[pair_info[1][i]])
                    not_oriented.remove(scanner)
                    break
                elif (scanner == pair_info[0][i][0]) and (pair_info[0][i][1] in orient_transform):
                    orient_transform[scanner] = np.matmul(orient_transform[pair_info[0][i][1]], inv_rotations[pair_info[1][i]])
                    not_oriented.remove(scanner)
                    break

     
    oriented = defaultdict(int)
    for i in range(len(dictionary)):
        oriented[i] = [np.matmul(orient_transform[i], beacon) for beacon in dictionary[i]]
    
    return oriented
```

```python
def shift(dictionary):
    '''Shifting (CORRECTLY oriented) scanners (output from orient() function) 
    such that the origin is the same as scanner 0. Returns the relative 
    position of each scanner to scanner 0.'''
    
    dictionary = orient(dictionary)
    
    shift_transform = defaultdict(int)
    not_shifted = list(range(1,len(dictionary)))
    
    pair_info = get_pairs(dictionary)
    
    # Initial conditions (Scanners who are 1 shift away from scanner 0)
    shift_transform[0] = np.zeros((3))

    for i in range(len(get_pairs(dictionary)[0])):
        if pair_info[0][i][0] == 0:
            not_shifted.remove(pair_info[0][i][1])
            shift_transform[pair_info[0][i][1]] = pair_info[2][i]
            
            
    while len(not_shifted) != 0:
         for scanner in not_shifted:
            for i in range(len(pair_info[0])):
                if (scanner == pair_info[0][i][1]) and (pair_info[0][i][0] in shift_transform):
                    shift_transform[scanner] = np.add(shift_transform[pair_info[0][i][0]], pair_info[2][i])
                    not_shifted.remove(scanner)
                    break
                elif (scanner == pair_info[0][i][0]) and (pair_info[0][i][1] in shift_transform):
                    shift_transform[scanner] = np.subtract(shift_transform[pair_info[0][i][1]], pair_info[2][i])
                    not_shifted.remove(scanner)
                    break
    
    
    
    
    return shift_transform
```

```python
def unique(dictionary):
    '''Takes in a data set. Calls the orient function to orient the data set.
    After the data set is oriented, call the shift function to get the shift
    transformation matrix. Apply that transformation matrix to the respective
    scanners. Returns the number of unique beacons 
    '''
    oriented_dictionary = orient(dictionary)
    shift_transform = shift(dictionary)
    beacons = set()
    for scanner in range(len(dictionary)):
        for beacon in range(len(dictionary[scanner])):
            beacon = np.add(shift_transform[scanner], oriented_dictionary[scanner][beacon])
            beacon = [int(x) for x in beacon]
            beacons.add(str(beacon))
    
    return len(beacons)
```

```python
def manhattan(scanners):
    max_dist = 0
    for scanner in range(len(scanners)):
        for pair in range(len(scanners)):
            dist = sum([abs(x) for x in scanners[scanner]-scanners[pair]])
            if dist > max_dist:
                max_dist = dist
                
    return int(max_dist)
```
